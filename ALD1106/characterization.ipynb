```python
import numpy as np
import matplotlib.pyplot as plt

# Custom settings
plt.style.use('classic')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['axes.edgecolor'] = 'black'
plt.rcParams['axes.linewidth'] = 1.2
plt.rcParams['savefig.facecolor'] = 'white'
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial']
plt.rcParams['mathtext.fontset'] = 'cm'
plt.rcParams['figure.dpi'] = 100
```

```python
# IDS - VDS data
data1 = np.loadtxt("./Data-IDS-VDS/20262004006.dat")
data2 = np.loadtxt("./Data-IDS-VDS/20262004007.dat")
data3 = np.loadtxt("./Data-IDS-VDS/20262004008.dat")
data4 = np.loadtxt("./Data-IDS-VDS/20262004009.dat")
data5 = np.loadtxt("./Data-IDS-VDS/20262004010.dat")
# IDS - VGS data
data6 = np.loadtxt("./Data-IDS-VGS/20262004011.dat")
data7 = np.loadtxt("./Data-IDS-VGS/20262004012.dat")
data8 = np.loadtxt("./Data-IDS-VGS/20262004013.dat")
data9 = np.loadtxt("./Data-IDS-VGS/20262004014.dat")
data10 = np.loadtxt("./Data-IDS-VGS/20262004015.dat")
# IGS - VGS data
data11 = np.loadtxt("./Data-IGS-VGS/20262004015.dat")
```

```python
datasets_IDS_VDS = [
    ("10 V", data5, 'red', 'o'),
    ("8 V", data4, 'blue', 's'),
    ("6 V", data3, 'green', '^'),
    ("4 V", data2, 'orange', 'D'),
    ("2 V", data1, 'purple', 'v')
]
for label, data, color, marker in datasets_IDS_VDS:
    V_DS = data[:, 0]
    I_DS = data[:, 1]
    plt.plot(V_DS, I_DS*1e3, lw=2, label=label, color=color, marker=marker, linestyle='-', markevery=40)

plt.xlabel(r"$V_\text{DS}$ (V)")
plt.ylabel(r"$I_\text{DS}$ (mA)")
plt.title("Output Characteristics of MOSFET")
plt.legend(frameon=True, loc='upper left', numpoints=1, fontsize=12)
plt.show()
```

```python
datasets_IDS_VGS = [
    ("8 V", data6, 'red', 'o'),
    ("5 V", data7, 'blue', 's'),
    ("2 V", data8, 'green', '^'),
    ("1 V", data9, 'orange', 'D'),
    ("0.2 V", data10, 'purple', 'v')
]
for label, data, color, marker in datasets_IDS_VGS:
    V_GS = data[:, 0]
    I_DS = data[:, 1]
    plt.plot(V_GS, I_DS*1e3, lw=2, label=label, color=color, marker=marker, linestyle='-', markevery=40)

plt.xlabel(r"$V_\text{GS}$ (V)")
plt.ylabel(r"$I_\text{DS}$ (mA)")
plt.title("Transfer Characteristics of MOSFET")
plt.legend(frameon=True, loc='upper left', numpoints=1, fontsize=12)
plt.show()
```

```python
for label, data, color, marker in datasets_IDS_VGS:
    V_GS = data[:, 0]
    I_DS = data[:, 1]
    plt.semilogy(V_GS, I_DS, lw=2, label=label, color=color, marker=marker, linestyle='-', markevery=50)

plt.xlabel(r"$V_\text{GS}$ (V)")
plt.ylabel(r"$I_\text{DS}$ (A)")
plt.title("Transfer Characteristics of MOSFET")
plt.legend(frameon=True, loc='lower right', numpoints=1)
plt.show()
```

```python
for label, data, color, marker in [("IGS", data11, 'red', '^')]:
    V_GS = data[:, 0]
    I_GS = data[:, 1]
    plt.plot(V_GS, I_GS*1e6, lw=2, label=label, color=color, marker=marker, linestyle='-', markevery=10)

plt.xlabel(r"$V_\text{GS}$ (V)")
plt.ylabel(r"$I_\text{GS}$ (uA)")
plt.title("Gate Leakage Current of MOSFET")
plt.legend(frameon=True, loc='upper left', numpoints=1, fontsize=12)
plt.show()
```
