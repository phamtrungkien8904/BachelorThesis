```python
import numpy as np
import matplotlib.pyplot as plt

# Custom settings
plt.style.use('classic')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['axes.edgecolor'] = 'black'
plt.rcParams['axes.linewidth'] = 1.2
plt.rcParams['savefig.facecolor'] = 'white'
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial']
plt.rcParams['mathtext.fontset'] = 'cm'
plt.rcParams['figure.dpi'] = 100
```

```python
# IDS - VDS data
data1 = np.loadtxt("./Data-IDS-VDS/20262004008.dat")
data2 = np.loadtxt("./Data-20262805-output/20262805003.dat")
data3 = np.loadtxt("./Data-20262805-output/20262805012.dat")

# IDS - VGS data
data4 = np.loadtxt("./Data-IDS-VGS/20262004012.dat")
data5 = np.loadtxt("./Data-20262805-transfer/20262805001.dat")
data6 = np.loadtxt("./Data-20262805-transfer/20262805011.dat")

```

```python
datasets_IDS_VDS = [
    ("373 K", data2, 'blue', 's', 20),
    ("295 K", data1, 'green', '^', 40),
    ("150 K", data3, 'orange', 'D', 20),
]
for label, data, color, marker, markevery in datasets_IDS_VDS:
    V_DS = data[:, 0]
    I_DS = data[:, 1]
    plt.plot(V_DS, I_DS*1e3, lw=2, label=r"$T \approx $" + label, color=color, marker=marker, linestyle='-', markevery=markevery, markeredgecolor='white', markeredgewidth=1, markersize=7)

plt.xlabel(r"Drain Voltage $V_\text{DS}$ (V)", fontsize=14)
plt.ylabel(r"Drain Current $I_\text{DS}$ (mA)", fontsize=14)
plt.xlim(0, 10)
plt.ylim(0, 20)
plt.title(r"Output Characteristics of ALD1106 ($V_\text{GS} = 6$ V)", fontsize=15)
plt.legend(frameon=True, loc='upper left', numpoints=1, fontsize=15)
# plt.savefig("output_characteristics.eps", format='eps', bbox_inches='tight')
plt.show()
```

```python
datasets_IDS_VGS = [
    ("373 K", data5, 'blue', 's', 10),
    ("295 K", data4, 'green', '^', 50),
    ("150 K", data6, 'orange', 'D', 10),
]
for label, data, color, marker, markevery in datasets_IDS_VGS:
    V_GS = data[:, 0]
    I_DS = data[:, 1]
    plt.plot(V_GS, I_DS*1e3, lw=2, label=r"$T \approx $" + label, color=color, marker=marker, linestyle='-', markevery=markevery,markeredgecolor="white", markeredgewidth=1, markersize=7)

plt.xlabel(r"Gate Voltage $V_\text{GS}$ (V)", fontsize=14)
plt.ylabel(r"Drain Current $I_\text{DS}$ (mA)", fontsize=14)
plt.ylim(0, 10)
plt.xlim(0, 5)
plt.title(r"Transfer Characteristics of ALD1106 ($V_\text{DS} = 5$ V)", fontsize=15)
plt.legend(frameon=True, loc='upper left', numpoints=1, fontsize=15)
# plt.savefig("transfer_characteristics.eps", format='eps', bbox_inches='tight')
plt.show()
```
