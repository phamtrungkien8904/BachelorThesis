```python
import numpy as np
import matplotlib.pyplot as plt

# Custom settings
plt.style.use('classic')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['axes.edgecolor'] = 'black'
plt.rcParams['axes.linewidth'] = 1.2
plt.rcParams['savefig.facecolor'] = 'white'
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial']
plt.rcParams['mathtext.fontset'] = 'cm'
plt.rcParams['figure.dpi'] = 100
```

```python
data = np.loadtxt('Data-20260605/20260506-0002.csv', delimiter=',', skiprows=1)
time = data[:, 0]
V_in = data[:, 1]
V_out = data[:, 2]

window_size = 10
window = np.ones(window_size) / window_size

V_in_smooth = np.convolve(V_in, window, mode='same')
V_out_smooth = np.convolve(V_out, window, mode='same')

plt.plot(time, V_in_smooth, label='Input Voltage', color='blue', lw=1.5)
plt.plot(time, V_out_smooth, label='Output Voltage', color='red', lw=1.5)
plt.xlabel('Time (us)', fontsize=12)
plt.ylabel('Voltage (V)', fontsize=12)
plt.xlim(-400, 400)
plt.ylim(-1, 1)
plt.title('FET Amplifier Response', fontsize=14)
plt.legend()
plt.show()
```
