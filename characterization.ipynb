```python
import numpy as np
import matplotlib.pyplot as plt

# Custom settings
plt.style.use('classic')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['savefig.facecolor'] = 'white'
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman', 'Times', 'DejaVu Serif']
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['figure.dpi'] = 100
```

```python
# IDS - VDS data
data1 = np.loadtxt("./Data_KP_IDS_VDS/20261504001.dat")
data2 = np.loadtxt("./Data_KP_IDS_VDS/20261504002.dat")
data3 = np.loadtxt("./Data_KP_IDS_VDS/20261504003.dat")
# IDS - VGS data
data4 = np.loadtxt("./Data_KP_IDS_VGS/20261504001.dat")
data5 = np.loadtxt("./Data_KP_IDS_VGS/20261504002.dat")
# IGS - VGS data
data6 = np.loadtxt("./Data_KP_IGS_VGS/20261504000.dat")
data7 = np.loadtxt("./Data_KP_IGS_VGS/20261504001.dat")
data8 = np.loadtxt("./Data_KP_IGS_VGS/20261504002.dat")
data9 = np.loadtxt("./Data_KP_IGS_VGS/20261504003.dat")
data10 = np.loadtxt("./Data_KP_IGS_VGS/20261504004.dat")
```

```python
datasets_IDS_VDS = [
    ("5 V", data1, 'red', 'o'),
    ("2 V", data2, 'blue', 's'),
    ("8 V", data3, 'green', '^'),
]
for label, data, color, marker in datasets_IDS_VDS:
    V_DS = data[:, 0]
    I_DS = data[:, 1]
    plt.plot(V_DS, I_DS, lw=2, label=label, color=color, marker=marker, linestyle='-', markevery=5)

plt.xlabel(r"$V_\text{DS}$ (V)")
plt.ylabel(r"$I_\text{DS}$ (A)")
plt.title("Output Characteristics of MOSFET")
plt.legend(frameon=True, loc='upper left', numpoints=1)
plt.show()

for label, data, color, marker in datasets_IDS_VDS:
    V_DS = data[:, 0]
    R_DS = data[:, 5]
    plt.semilogy(V_DS, R_DS, lw=2, label=label, color=color, marker=marker, linestyle='-', markevery =5)

plt.xlabel(r"$V_\text{DS}$ (V)")
plt.ylabel(r"$R_\text{DS}$ ($\Omega$)")
plt.title("Drain Source Resistance of MOSFET")
plt.legend(frameon=True, loc='upper left', numpoints=1)
plt.show()
```

```python
datasets_IDS_VGS = [
    ("5 V", data5, 'blue'),
    ("2 V", data4, 'red'),
]
for label, data, color in datasets_IDS_VGS:
    V_GS = data[:, 0]
    I_DS = data[:, 1]
    plt.plot(V_GS, I_DS, lw=2, label=label, color=color, marker='^', linestyle='-', markevery=5)

plt.xlabel(r"$V_\text{GS}$ (V)")
plt.ylabel(r"$I_\text{DS}$ (A)")
plt.title("Transfer Characteristics of MOSFET")
plt.legend(frameon=True, loc='upper left', numpoints=1)
plt.show()
```

```python
datasets_IGS_VGS = [
    ("0 V", data6, 'red'),
    ("1 V", data7, 'blue'),
    ("2 V", data8, 'green'),
    ("3 V", data9, 'orange'),
    ("4 V", data10, 'purple'),
]
for label, data, color in datasets_IGS_VGS:
    V_GS = data[:, 0]
    I_GS = data[:, 1]
    plt.plot(V_GS, I_GS, lw=2, label=label, color=color, marker='^', linestyle='-', markevery=5)

plt.xlabel(r"$V_\text{GS}$ (V)")
plt.ylabel(r"$I_\text{GS}$ (A)")
plt.title("Leakage Characteristics of MOSFET")
plt.legend(frameon=True, loc='upper left', numpoints=1)
plt.show()
```

```python
data_m = np.loadtxt("./Mobility/20260318002.dat")
V_G4 = data_m[:, 0]
I_G4 = data_m[:, 1]
I_34 = data_m[:, 2]
V_34 = data_m[:, 3]
V_14 = data_m[:, 4]
V_24 = data_m[:, 5]
V_12 = data_m[:, 6]
sigma = data_m[:, 7]
n_2D = data_m[:, 8]
mu = -data_m[:, 9]
```

```python
plt.plot(V_G4, mu, lw=2, label=r"$\mu$", color='blue', marker='o', linestyle='-', markevery=5)
plt.xlabel(r"$V_\text{G4}$ (V)")
plt.ylabel(r"Mobility $\mu$ (cm$^2$/V$\cdot$s)")
plt.title("Mobility of MOSFET")
plt.xlim(-40,0)
plt.ylim(-10, 10)
plt.legend(frameon=True, loc='upper left', numpoints=1)
```

```python
# Parameters
C = 2.6e-9
A = 33e-6
V_T = 5.0

n_2D_theory = C / A * (V_G4 - V_T)
sigma_theory= np.log(2)/np.pi * (I_34 / V_12) 
mu_corr = -sigma_theory / n_2D_theory * 10000

plt.plot(V_G4, mu_corr, lw=2, label=r"$\mu$ (theory)", color='red', marker='s', linestyle='-', markevery=1)
plt.xlabel(r"$V_\text{G4}$ (V)")
plt.ylabel(r"Mobility $\mu$ (cm$^2$/V$\cdot$s)")
plt.xlim(-40,0)
plt.ylim(-10, 10)
plt.title("Mobility of MOSFET")
plt.legend(frameon=True, loc='upper left', numpoints=1)
```
