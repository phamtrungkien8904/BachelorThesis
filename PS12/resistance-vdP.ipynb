```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Custom settings
plt.style.use('classic')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['axes.edgecolor'] = 'black'
plt.rcParams['axes.linewidth'] = 1.2
plt.rcParams['savefig.facecolor'] = 'white'
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial']
plt.rcParams['mathtext.fontset'] = 'cm'
plt.rcParams['figure.dpi'] = 100

# IDS - VGS data
data = np.loadtxt("./Data-vdP/20262304005.dat")
V_GS = data[:, 0]  # VGS values
I_DS = data[:, 2]  # IDS values
V_DS = data[:, 3]  # VDS values
sigma = data[:, 7]

V_C = 0.5*(data[:, 4] + data[:, 5])
R_DS = V_DS / I_DS  # Calculate resistance
V_T = -10.1
C =	2.6e-9	# F
A = 45e-6	# m^2
Ci = C/A # F/m^2
mu = 3.6e-4	# m^2/Vs

x = V_GS - V_T - V_C

plt.plot(-x, R_DS*1e-6, lw=2, label=r'Data $R_\text{34}$', color='red', marker='o', linestyle='-', markevery=40)
plt.xlabel(r"$-(V_\text{G4}- V_\text{T} - V_\text{C})$ (V)")
plt.ylabel(r"$R_\text{34}$ (M$\Omega$)")
plt.xlim(5,30)
plt.ylim(50, 150)
plt.legend(frameon=True, loc='upper right', numpoints=1)
plt.title("Resistance vs Gate Voltage")
plt.show()
```

```python
def func(x, a, b):
    return a + b / x

# Use only finite points in the requested fit window: 5 <= -x <= 30
mask = np.isfinite(x) & np.isfinite(R_DS) & (-x >= 15) & (-x <= 30)
x_fit = x[mask]
R_fit_data = R_DS[mask]

# Target b value corresponding to mu = 3 cm^2/Vs
b_target = -np.log(2) / (np.pi * Ci * 1e-4 * 3)
a0 = np.median(R_fit_data)

# Fit Rc + b/x to the selected x-range
popt, pcov = curve_fit(
    func,
    x_fit,
    R_fit_data,
    p0=(a0, b_target),
    bounds=([-np.inf, -np.inf], [np.inf, 0]),
)
R_fit = func(x_fit, *popt)
a, b = popt

# Coefficient of determination
ss_res = np.sum((R_fit_data - R_fit) ** 2)
ss_tot = np.sum((R_fit_data - np.mean(R_fit_data)) ** 2)
r2 = 1 - (ss_res / ss_tot)

# 1-sigma parameter uncertainties from covariance matrix
perr = np.sqrt(np.diag(pcov))
a_err, b_err = perr



R_fit = func(x_fit, a, b)


print(f"a = {a:.6g} ± {a_err:.3g}")
print(f"b = {b:.6g} ± {b_err:.3g}")
print(f"R^2 = {r2:.6f}")
print("--------------------------------------")
print(f"Rc = ({a*1e-6:.6g} ± {a_err*1e-6:.3g}) MΩ")

# x_test = np.linspace(-30, -5, 100)
# R_test = func(x_test, a, b_target)  # Reference curve with mu=3 cm^2/Vs

plt.plot(-x, R_DS*1e-6, lw=2, label=r'Data $R_\text{34}$', color='red', marker='o', linestyle='-', markevery=20)
plt.plot(-x_fit, R_fit*1e-6, lw=2, label='Fit: $R_C + b/x$', color='blue', linestyle='--')
# plt.plot(-x_test, R_test*1e-6, lw=2, label='Test Fit', color='green', linestyle='--')
plt.xlabel(r"$-(V_\text{G4}- V_T - V_\text{34}/2)$ (V)")
plt.ylabel(r"$R_\text{34}$ (M$\Omega$)")
plt.xlim(10,30)
plt.ylim(0, 200)
plt.legend(frameon=True, loc='upper right', numpoints=1)
plt.title("Total Resistance vs Gate Voltage (PS12)")
# plt.savefig("total_PS12.eps", format='eps', bbox_inches='tight')
plt.show()
```

```python
R_channel = 1/sigma

plt.plot(-x, R_channel*1e-6, lw=2, label=r'Calculated $R_\text{channel}$', color='red', marker='^', linestyle='None', markevery=1)
plt.xlabel(r"$-(V_\text{G4}- V_T - V_C)$ (V)")
plt.ylabel(r"$R_\text{channel}$ (M$\Omega$)")
plt.xlim(10,30)
plt.ylim(0, 10)
plt.legend(frameon=True, loc='upper right', numpoints=1)
plt.title("Channel Resistance vs Gate Voltage (PS12)")
plt.show()
```
